```python
sara = ventas[ventas.Vendedor.str.contains('Bond')]
sara
```

```python
sara.Artículo.value_counts().plot();
```

```python
keys = []
values = []
for key, value in sara.Categoría.value_counts().to_dict().items():
    print(key, ' - ' ,value)
    keys.append(key)
    values.append(value)

plt.plot(keys,values, 'r>:');
```
